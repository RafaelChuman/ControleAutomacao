```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import animation, rc

plt.rcParams['figure.figsize'] = (8,8) # tamanho do gráfico
plt.rcParams['font.size'] = 14 # tamanho da letraimport matplotlib.pyplot as plt
```

```python
# Animação do Delta de Dirac
fig, ax = plt.subplots() 

# intervalo no eixo x
x = np.linspace(-10, 10, 1000)
i = 1
delta = (1/(np.sqrt(np.pi)*i))*np.exp(-x**2/i**2) # função
line, = ax.plot(x, delta)

# definição para os eixo ficarem fixos
def init():
    ax.set_xlim(-5, 5)
    ax.set_ylim(-2, 30)
    return line,

# animação da função
def animate(i):
    a = i
    line.set_ydata((1/(np.sqrt(np.pi)*a))*np.exp(-x**2/a**2))  # update the data.
    return line,

# https://matplotlib.org/stable/api/_as_gen/matplotlib.animation.FuncAnimation.html
anim = animation.FuncAnimation(
    fig, animate, np.linspace(1, 0.01, 10), init_func=init, blit=True)

rc('animation', html='jshtml')
anim  



```
