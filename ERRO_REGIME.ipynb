```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl
from control import (TransferFunction)

plt.rcParams['figure.figsize'] = (10,10) # tamanho do gráfico
plt.rcParams['font.size'] = 14 # tamanho da letraimport matplotlib.pyplot as plt
```

```python
# Comparação entre as respostas temporais para sistemas do tipo k=0 e do tipo k=1

# malha aberta
L1 = TransferFunction(1, [1,1]) # tipo k=0
L2 = TransferFunction(1, [1,1,0]) # tipo k=1

# malha fechada
plant_tf1 = ctl.feedback(L1, 1)
plant_tf2 = ctl.feedback(L2, 1)

t_ = np.linspace(0, 10, 1001)
u_ = step_input = t_>0 # entrada degrau

plt.figure(1)
tout1, yout1 = ctl.step_response(plant_tf1, T=t_)
tout2, yout2 = ctl.step_response(plant_tf2, T=t_)
plt.plot(t_, u_, 'k-', linewidth=1.5, label='entrada')
plt.plot(tout1, yout1, 'b--', linewidth=1.5, label='resposta tipo k=0')
plt.plot(tout2, yout2, 'r-.', linewidth=1.5, label='resposta tipo k=1')
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()

```
