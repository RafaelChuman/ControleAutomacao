```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import animation, rc

plt.rcParams['figure.figsize'] = (6,6) # tamanho do gráfico
plt.rcParams['font.size'] = 14 # tamanho da letraimport matplotlib.pyplot as plt
```

```python
# Caminho de s não envolve polo nem zero
fig1, ax = plt.subplots()
xdata, ydata = [], [] # dados do caminho do s
pxdata, pydata = [], [] # polo -1
zxdata, zydata = [], [] # zero -2
gxdata, gydata = [], [] # dados do caminho do G(s)=(s+2)/(s+1)
ln, = plt.plot([], [], 'r') # caminho s
pln, = plt.plot([], [], 'k*') # polo
zln, = plt.plot([], [], 'ko') # zero 
gln, = plt.plot([], [], 'b') # caminho G

def init():
    ax.set_xlim(-2, 4)
    ax.set_ylim(-2, 2)
    return ln,

def update(frame):
    # polo em (-1,0)
    pxdata.append(-1)
    pydata.append(0)
    pln.set_data(pxdata,pydata)
    # zero em (-2,0)
    zxdata.append(-2)
    zydata.append(0)
    zln.set_data(zxdata,zydata)
    # caminho percorrido
    s = 2 + np.sin(frame) + 1j*np.cos(frame) # horário
    xdata.append(np.real(s))
    ydata.append(np.imag(s))
    ln.set_data(xdata,ydata)
    # caminho percorrido da planta com poles em -1 e zero em -2
    g = (s+2)/(s+1)
    gxdata.append(np.real(g))
    gydata.append(np.imag(g))
    gln.set_data(gxdata,gydata)
    return ln,pln,zln,gln

anim = FuncAnimation(fig1, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
rc('animation', html='jshtml')
anim  
#plt.show()
```

```python
# Caminho de s envolve um polo (-1)
plt.figure(2)

fig2, ax = plt.subplots()
x2data, y2data = [], [] # caminho s
px2data, py2data = [], []
zx2data, zy2data = [], []
gx2data, gy2data = [], [] # caminho g
ln, = plt.plot([], [], 'r')
pln, = plt.plot([], [], 'k*')
zln, = plt.plot([], [], 'ko')
gln, = plt.plot([], [], 'b')

def init():
    ax.set_xlim(-2, 4)
    ax.set_ylim(-2, 2)
    return ln,

def update(frame):
    # polo em (-1,0)
    px2data.append(-1)
    py2data.append(0)
    pln.set_data(px2data,py2data)
    # zero em (-2,0)
    zx2data.append(-2)
    zy2data.append(0)
    zln.set_data(zx2data,zy2data)
    # caminho percorrido
    s = - 0.5 + np.sin(frame)+1j*np.cos(frame) # sentido horario
    x2data.append(np.real(s))
    y2data.append(np.imag(s))
    ln.set_data(x2data,y2data)
    # caminho percorrido da planta com poles em -1 e zero em -2
    g = (s+2)/(s+1) # sentido anti-horario
    gx2data.append(np.real(g))
    gy2data.append(np.imag(g))
    gln.set_data(gx2data,gy2data)
    return ln,pln,zln,gln

anim = FuncAnimation(fig2, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
rc('animation', html='jshtml')
anim
```
