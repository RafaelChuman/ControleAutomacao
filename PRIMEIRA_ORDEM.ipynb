```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl
from control import (TransferFunction)

plt.rcParams['figure.figsize'] = (10,10) # tamanho do gráfico
plt.rcParams['font.size'] = 14 # tamanho da letraimport matplotlib.pyplot as plt

# Entrada Degrau
# neste arquivo usamos 2 opções step_response e o forced_response usando heaviside

y0 = [0] # condição inicial nula
t = np.linspace(0, 8, 1001) # vetor tempo

# Função Transferência
a = 1
plant_tf = TransferFunction(a, [1,a])
print("TF = \n", plant_tf)
```

```python
# Constante de Tempo

plt.figure(1)
xout, yout = ctl.step_response(plant_tf, T=t)
plt.plot(t, yout, 'k', linewidth=1.5, label='posição')
plt.plot(1/a, 1 - np.exp(-1), 'b*', markersize=12)
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()
```

```python
# Tempo de Subida

# ver https://numpy.org/doc/stable/reference/generated/numpy.heaviside.html
u = np.heaviside(t, 0) 

plt.figure(2)
xout, yout = ctl.forced_response(plant_tf, U=u, T=t)
plt.plot(t, yout, 'k', linewidth=1.5, label='posição')
plt.plot(-(1/a) * np.log(1 - 0.1), 0.1, 'b*', markersize=12)
plt.plot(-(1/a) * np.log(1 - 0.9), 0.9, 'b*', markersize=12)
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()
```

```python
# Tempo de acomodação Ta (2%)
plt.figure(3)
xout, yout = ctl.forced_response(plant_tf, U=u, T=t)
plt.plot(t, yout, 'k', linewidth=1.5, label='posição')
plt.plot(4/a, 1 - np.exp(-4), 'g*', markersize=12)
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()
```

```python
a1 = 10
a2 = 50

plant_tf1 = TransferFunction(a1, [1,a1])
plant_tf2 = TransferFunction(a2, [1,a2])

print("TF1 =\n", plant_tf1)
print("TF2 =\n", plant_tf2)

t_ = np.linspace(0, 0.6, 1001)
u_ = np.heaviside(t_, 0) # ver https://numpy.org/doc/stable/reference/generated/numpy.heaviside.html

plt.figure(4)
xout1, yout1 = ctl.forced_response(plant_tf1, U=u_, T=t_)
xout2, yout2 = ctl.forced_response(plant_tf2, U=u_, T=t_)
plt.plot(t_, yout1, 'k', linewidth=1.5, label='a=10')
plt.plot(t_, yout2, 'k--', linewidth=1.5, label='a=50')
plt.plot(1/a1, 1 - np.exp(-1), 'r*', markersize=12)
plt.plot(1/a2, 1 - np.exp(-1), 'r*', markersize=12)
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()


plt.figure(5)
ctl.pzmap(plant_tf1, plot=True, grid=True, title='Pole Zero Map')
ctl.pzmap(plant_tf2, plot=True, grid=True, title='Pole Zero Map')
plt.show()
```
