```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl
from control import (TransferFunction, step_response, bode_plot,
                     feedback, sisotool)

plt.rcParams['figure.figsize'] = (10,10) # tamanho do gr√°fico
plt.rcParams['font.size'] = 16 # tamanho da letra
```

```python
# Compensador Avan√ßo de fase
K = 1
T = 1
# alpha < 1
# lead = TransferFunction(K * [T, 1],[alpha * T, 1]);

lead_01 = TransferFunction(K * [T, 1],[0.01 * T, 1])
lead_1  = TransferFunction(K * [T, 1],[0.1  * T, 1])
lead_5  = TransferFunction(K * [T, 1],[0.5  * T, 1])

w = np.linspace(0.1, 100, 1001)
plt.figure() # Diagrama de Bode dos compensadores Avan√ßo de fase  
mag, phase, omega = bode_plot(lead_01, omega=w, dB=True, linewidth = 2, linestyle='solid', color='black')
mag, phase, omega = bode_plot(lead_1, omega=w, dB=True, linewidth = 2, linestyle='dashed', color='red')
mag, phase, omega = bode_plot(lead_5, omega=w, dB=True, linewidth = 2, linestyle='dashdot', color='blue')
plt.legend([r"$\alpha = 0.01$", r"$\alpha = 0.1$", r"$\alpha = 0.5$"])
plt.show
```

```python
# Compensador Avan√ßo de fase similar ao PD
Kp = 1
Td = 1.1
# alpha < 1
# lead = TransferFunction(K * [T, 1], [alpha * T, 1]);

PD = TransferFunction(Kp * [Td, 1], 1)

w = np.linspace(0.1, 1000, 10001)
plt.figure() 
mag, phase, omega = bode_plot(PD, omega=w, dB=True, linewidth = 2, linestyle='solid', color='black')
plt.legend(["PD"])
plt.show
```

```python
# Compensador Atraso de fase
K = 1
T = 1
# alpha > 1
lag_5  = TransferFunction(K * [T, 1],[5  * T, 1]) # alpha = 5
lag_10 = TransferFunction(K * [T, 1],[10 * T, 1]) # alpha = 10
lag_50 = TransferFunction(K * [T, 1],[50 * T, 1]) # alpha = 50

w = np.linspace(0.1, 100, 1001)
plt.figure(2)  # Diagrama de Bode dos compensadores Atraso de fase
mag, phase, omega = bode_plot(lag_5, omega=w, dB=True, linewidth = 2, linestyle='solid', color='black')
mag, phase, omega = bode_plot(lag_10, omega=w, dB=True, linewidth = 2, linestyle='dashed', color='red')
mag, phase, omega = bode_plot(lag_50, omega=w, dB=True, linewidth = 2, linestyle='dashdot', color='blue')
plt.legend([r"$\alpha = 5$", r"\alpha = 10", r"$alpha = 50"])
plt.show
```

```python
# Compensador Atraso de fase similar ao PI
Kp = 1
Ti = 0.1
# alpha < 1
# lag = TransferFunction(K * [T, 1], [alpha * T, 1]);

PI = TransferFunction([Ti * Kp, Kp], [Ti, 0])

w = np.linspace(0.1, 1000, 10001)
plt.figure() 
mag, phase, omega = bode_plot(PI, omega=w, dB=True, linewidth = 2, linestyle='solid', color='black')
plt.legend(["PI"])
plt.show
```
