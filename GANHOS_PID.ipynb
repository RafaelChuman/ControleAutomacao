```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl
from control import (TransferFunction, bode_plot, step_response, feedback, pzmap)

plt.rcParams['figure.figsize'] = (10,10) # tamanho do gr√°fico
plt.rcParams['font.size'] = 16 # tamanho da letra
```

```python
# Ganho Proporcional

# malha aberta
L = TransferFunction(1, [1,1]) # 1/s+1

# malha fechada com aumento do ganho proprocional
sys1 = feedback(1*L, 1) # Kp = 1
sys2 = feedback(10*L, 1) # Kp = 10
sys3 = feedback(100*L, 1) # Kp = 100

w = np.linspace(0.1, 100, 1001)

plt.figure(1)  # Diagrama de Bode da Malha aberta
mag, phase, omega = bode_plot(1*L, omega=w, dB=True, ls='-')
mag, phase, omega = bode_plot(10*L, omega=w, dB=True, ls='--')
mag, phase, omega = bode_plot(100*L, omega=w, dB=True, ls='-.')
plt.legend(["K=1", "K=10", "K=100"])
plt.show

t = np.linspace(0, 2, 1001)

plt.figure(2) # Resposta ao Degrau Malha Fechada
tout, yout1 = step_response(sys1, t, X0=0)
tout, yout2 = step_response(sys2, t, X0=0)
tout, yout3 = step_response(sys3, t, X0=0)
plt.plot(tout, yout1, 'b', linewidth=1.5, label='K=1')
plt.plot(tout, yout2, 'k--', linewidth=1.5, label='K=10')
plt.plot(tout, yout3, 'g-.', linewidth=1.5, label='K=100')
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.show()
```

```python
# Ganho Derivativo

# malha aberta
L = TransferFunction([1, 0], [1,1]) 

# malha fecahda
sys1 = feedback(1*L, 1)
sys2 = feedback(10*L, 1)
sys3 = feedback(100*L, 1)

w = np.linspace(0.1, 100, 1001)

plt.figure(3)
mag, phase, omega = bode_plot(1*L, omega=w, dB=True, ls='-')
mag, phase, omega = bode_plot(10*L, omega=w, dB=True, ls='--')
mag, phase, omega = bode_plot(100*L, omega=w, dB=True, ls='-.')
plt.legend(["K=1", "K=10", "K=100"])
plt.show

t = np.linspace(0, 20, 1001)

plt.figure(4)
tout, yout1 = step_response(sys1, t, X0=0)
tout, yout2 = step_response(sys2, t, X0=0)
tout, yout3 = step_response(sys3, t, X0=0)
plt.plot(tout, yout1, 'b-', linewidth=1.5, label='K=1')
plt.plot(tout, yout2, 'k--', linewidth=1.5, label='K=10')
plt.plot(tout, yout3, 'g-.', linewidth=1.5, label='K=100')
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.show()


```
