```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl
from control import (TransferFunction)

plt.rcParams['figure.figsize'] = (10,10) # tamanho do gráfico
plt.rcParams['font.size'] = 14 # tamanho da letraimport matplotlib.pyplot as plt

y0 = [0,0] # condições iniciais nulas
```

```python
# fator de amortecimento >0
plant_tf1 = TransferFunction(9, [1,9,9])
print("TF1 =\n", plant_tf1)

t_ = np.linspace(0, 6, 1001) # vetor tempo

# resposta ao degrau
plt.figure(1)
tout1, yout1 = ctl.step_response(plant_tf1, T=t_)
plt.plot(tout1, yout1, 'k', linewidth=1.5)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()

# polos e zeros no plano real-imag
plt.figure(2)
ctl.pzmap(plant_tf1, plot=True, grid=True)
plt.show()
```

```python
# fator de amortecimento =1
plant_tf2 = TransferFunction(9, [1,6,9])
print("TF2 =\n", plant_tf2)

t_ = np.linspace(0, 6, 1001)# vetor tempo

# resposta ao degrau
plt.figure(3)
tout2, yout2 = ctl.step_response(plant_tf2, T=t_)
plt.plot(tout1, yout1, 'k', linewidth=1.5)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()

# polos e zeros no plano real-imag
plt.figure(4)
ctl.pzmap(plant_tf2, plot=True, grid=True)
plt.show()
```

```python
# fator de amortecimento entre 0 e 1
plant_tf4 = TransferFunction(9, [1,0,9])
print("TF4 =\n", plant_tf4)

t_ = np.linspace(0, 6, 1001)# vetor tempo

# Resposta ao Degrau
plt.figure(7)
tout4, yout4 = ctl.step_response(plant_tf4, T=t_)
plt.plot(tout4, yout4, 'k', linewidth=1.5)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()

# polos e zeros no plano real-imag
plt.figure(8)
ctl.pzmap(plant_tf4, plot=True, grid=True)
plt.show()


```

```python
# Comparação das respostas ao degrau para sistemas com fatores de amortecimento diferentes
plt.figure(9)
plt.plot(tout1, yout1, 'k', linewidth=2, label='fator > 1')
plt.plot(tout2, yout2, 'r:', linewidth=2, label='fator = 1')
plt.plot(tout3, yout3, 'b--', linewidth=2, label='0 < fator < 1')
plt.plot(tout4, yout4, 'g-.', linewidth=2, label='fator = 0')
plt.legend(loc='best', shadow=True, framealpha=1)
plt.grid(alpha=0.3)
plt.xlabel('tempo')
plt.ylabel('resposta')
plt.show()
```
